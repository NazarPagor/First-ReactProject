# ***Розпізнавання моделей одежі за допомогою Keras***

---



```python
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import utils
from tensorflow.keras.preprocessing import image
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline 
```

Підготовка данних


```python
#Cписок з назвами класів
classes = ['t-shirt/top', 'trouser', 'pullover', 'dress', 'coat', 'sandal', 'shirt', 'sneaker', 'bag', 'boot']
#Завантаження данних
(x_data, y_data), (x_test, y_test) = fashion_mnist.load_data()
# Та їхній перегляд
plt.figure(figsize=(10,10))
for i in range(100,150):
    plt.subplot(5,10,i-100+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_data[i], cmap=plt.cm.binary)
    plt.xlabel(classes[y_data[i]])
    
```

Перетворюємо та нормалізуємо данні  

```python
x_data = x_data.reshape(60000, 784)
x_data = x_data / 255 
x_test = x_test.reshape(10000, 784)
x_test = x_test / 255 
```

**Робота з правильними відповідями**

```python
print(y_data[0])
```

Перетворюємо мітки в формат one hot encoding 

```python
y_data = utils.to_categorical(y_data, 10)
y_test = utils.to_categorical(y_test, 10)
print(y_data[0])
```

# Робота з нейронною мережею

```python
model = Sequential()
model.add(Dense(800, input_dim=784, activation="relu"))
model.add(Dense(10, activation="softmax"))
model.compile(loss="categorical_crossentropy", optimizer="SGD", metrics=["accuracy"])
print(model.summary())
```

Навчаємо нашу мережу та запускаємо розпізнавання набору данних на якому навчили мережу

```python
model.fit(x_data, y_data, 
          batch_size=200, 
          epochs=100,  
          verbose=1)


predictions = model.predict(x_data)
```

Перевіряємо якість роботи 

```python
n = 0
plt.imshow(x_data[n].reshape(28, 28), cmap=plt.cm.binary)
plt.show()
#
print(predictions[n])
#Визначаємо номер класу зображення, який пропонує мережа
np.argmax(predictions[n])
#Друкуємо назву класу
classes[np.argmax(predictions[n])]

#Друкуємо номер класу правильної відповіді
np.argmax(y_data[n])
#Друкуємо назву класу правильної відповіді
classes[np.argmax(y_data[n])]
```

# ***Аналіз якості роботи нейронної мережі для розпізнавання моделей одежі в Keras***

```python
model.save('fashion_mnist_dense.h5')
```

Оцінка якості *навчання*

```python
#Перевірка якості роботи на наборі даних для тестування
scores = model.evaluate(x_test, y_test, verbose=1)
#Доля правильних відповідей на тестових даних, в відсотках
print("Доля правильних відповідей на тестових даних, в відсотках:", round(scores[1] * 100, 4))
```

Використовуємо мережу для розпізнавання предметів одежі

```python
n_rec = 493
plt.imshow(x_test[n_rec].reshape(28, 28), cmap=plt.cm.binary)
plt.show()


#Змінюємо розмірність зображення та нормалізуємо його

x = x_test[n_rec]
x = np.expand_dims(x, axis=0)

#Та запустимо результати розпізнавання

prediction = model.predict(x)

#Та виведемо їх
prediction
```

Перетворюємо результати із формату one hot encoding

```python
prediction = np.argmax(prediction)
print("Номер класу:", prediction)
print("Назва класу:", classes[prediction])
# Та видрукуємо нормальну відповідь
label = np.argmax(y_test[n_rec])
print("Номер класу:", label)
print("Назва класу:", classes[label])
```
